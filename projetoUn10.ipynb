```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

print("Por favor, fa√ßa upload do arquivo 'X_train.txt' (dados de treino).")
from google.colab import files
uploaded_train = files.upload()

print("Por favor, fa√ßa upload do arquivo 'y_train.txt' (atividades de treino).")
uploaded_labels = files.upload()

X_train = pd.read_csv('X_train.txt', delim_whitespace=True, header=None)
y_train = pd.read_csv('y_train.txt', delim_whitespace=True, header=None, names=['Activity'])

if X_train.isnull().sum().any():
    print("Aviso: Existem valores ausentes no conjunto de dados. Verifique e trate antes de prosseguir.")
else:
    print("Nenhum valor ausente detectado nos dados de treino.")

print("\nResumo dos dados de treino:")
print(X_train.describe())
print("\nFormato dos dados de treino:", X_train.shape)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_train_scaled)

print(f"\nVari√¢ncia explicada pelos 2 primeiros componentes: {sum(pca.explained_variance_ratio_)*100:.2f}%")

plt.figure(figsize=(10, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], s=10, c='gray', alpha=0.6)
plt.title('Dados projetados no espa√ßo 2D (PCA)')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.grid(True)
plt.show()

inertias = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(X_train_scaled)
    inertias.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(k_values, inertias, 'o-', color='blue')
plt.title('M√©todo do Cotovelo para Escolha de K')
plt.xlabel('N√∫mero de Clusters (K)')
plt.ylabel('In√©rcia')
plt.xticks(k_values)
plt.grid(True)
plt.show()

best_k = max(range(2, 11), key=lambda k: silhouette_score(X_train_scaled, KMeans(n_clusters=k, init='k-means++', random_state=42).fit_predict(X_train_scaled)))
print(f"\nN√∫mero √≥timo de clusters com base no Silhouette Score: {best_k}")

kmeans = KMeans(n_clusters=best_k, init='k-means++', random_state=42)
clusters = kmeans.fit_predict(X_train_scaled)

silhouette_avg = silhouette_score(X_train_scaled, clusters)
print(f"\nSilhouette Score para K={best_k}: {silhouette_avg:.2f}")

plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=clusters, palette='Set2', s=50, alpha=0.7)
plt.title(f'Clusters formados com K-means (K={best_k})')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend(title='Clusters', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()

cluster_activity = pd.DataFrame({'Cluster': clusters, 'Activity': y_train['Activity']})
cluster_summary = cluster_activity.groupby('Cluster')['Activity'].apply(lambda x: x.mode()[0])
print("\nAtividades principais em cada cluster:")
print(cluster_summary)

print("\nO modelo K-means foi aplicado com sucesso. Use as visualiza√ß√µes e m√©tricas para avaliar os resultados.")
```
